```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd
```

```python
# 加载作者的npy文件和坐标轴的npy文件
authors_vectors = np.load('author_ebd_2d.npy')
authors_vectors_100d = np.load('author_ebd_100d.npy')
axis_vector_hard = np.load('hard_2d.npy')
axis_vector_sciences = np.load('sciences_2d.npy')
axis_vector_theory = np.load('theory_2d.npy')
axis_vector_quantity = np.load('quantity_2d.npy')
axis_vector_hard_100d = np.load('hard_100d.npy')
axis_vector_sciences_100d = np.load('sciences_100d.npy')
axis_vector_theory_100d = np.load('theory_100d.npy')
axis_vector_quantity_100d = np.load('quantity_100d.npy')
```

```python
# 计算坐标轴向量的单位向量
axis_unit_vector_hard = axis_vector_hard / np.linalg.norm(axis_vector_hard)
axis_unit_vector_sciences = axis_vector_sciences / np.linalg.norm(axis_vector_sciences)
axis_unit_vector_theory = axis_vector_theory / np.linalg.norm(axis_vector_theory)
axis_unit_vector_quantity = axis_vector_quantity / np.linalg.norm(axis_vector_quantity)
axis_unit_vector_hard_100d = axis_vector_hard_100d / np.linalg.norm(axis_vector_hard_100d)
axis_unit_vector_sciences_100d = axis_vector_sciences_100d / np.linalg.norm(axis_vector_sciences_100d)
axis_unit_vector_theory_100d = axis_vector_theory_100d / np.linalg.norm(axis_vector_theory_100d)
axis_unit_vector_quantity_100d = axis_vector_quantity_100d / np.linalg.norm(axis_vector_quantity_100d)
```

```python
# 计算每个作者向量在坐标轴向量方向上的投影
projections_hard = np.dot(authors_vectors, axis_unit_vector_hard)
# 计算投影后的坐标
x_hard = projections_hard * axis_unit_vector_hard[0]
y_hard = projections_hard * axis_unit_vector_hard[1]
```

```python
projections_sciences = np.dot(authors_vectors, axis_unit_vector_sciences)
x_sciences = projections_sciences * axis_unit_vector_sciences[0]
y_sciences = projections_sciences * axis_unit_vector_sciences[1]
```

```python
projections_theory = np.dot(authors_vectors, axis_unit_vector_theory)
x_theory = projections_theory * axis_unit_vector_theory[0]
y_theory = projections_theory * axis_unit_vector_theory[1]
```

```python
projections_quantity = np.dot(authors_vectors, axis_unit_vector_quantity)
x_quantity = projections_quantity * axis_unit_vector_quantity[0]
y_quantity = projections_quantity * axis_unit_vector_quantity[1]
```

```python
projections_hard_100d = np.dot(authors_vectors_100d, axis_unit_vector_hard_100d)
x_hard_100d_1 = projections_hard_100d * axis_unit_vector_hard_100d[0]
x_hard_100d_2 = projections_hard_100d * axis_unit_vector_hard_100d[1]
x_hard_100d_3 = projections_hard_100d * axis_unit_vector_hard_100d[2]
x_hard_100d_4 = projections_hard_100d * axis_unit_vector_hard_100d[3]
...
x_hard_100d_100 = projections_hard_100d * axis_unit_vector_hard_100d[99]
```
